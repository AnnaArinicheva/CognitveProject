```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

def generate_data():
    t = np.linspace(0, 1, 100)
    data = np.sin(2 * np.pi * 5 * t) + np.random.randn(100) * 0.5
    return t, data

def smooth_data(t, data):
    filtered_data = signal.savgol_filter(data, 51, 3)
    return t, filtered_data

def create_plot(t, data):
    # Создаем график
    plt.figure()
    plt.plot(t, data, label='Original Data')
    plt.title('Generated Data')
    plt.xlabel('Time')
    plt.ylabel('Amplitude')
    plt.legend()
    plt.grid(True)
    plt.show()

def process_and_plot():
    t, data = generate_data()
    t, smoothed_data = smooth_data(t, data)
    plot_path = create_plot(t, smoothed_data)
    return plot_path

t, data = generate_data()
t, filtered_data = smooth_data(t, data)
create_plot(t, data)
```
